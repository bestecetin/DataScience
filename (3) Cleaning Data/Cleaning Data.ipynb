<h3>MISSING DATA and TESTING WITH ASSERT</h3> <br>
missing data'ya sahipsek neler yapabiliriz? <br>
<li>öylece bırakabiliriz, belki de gerçekten öyle bir değeri yoktur.</li>
<li>dropna( ) ile verisetinden çıkarabiliriz.</li>
<li>fillna( ) ile NaN doldurabiliriz.</li>
<li>Veri eğer sayısalsa diğer verilerin ortalamasına bakıp o değerle doldurabilirdik.</li>
Assert statement: yaptıgımız şeyi kontrol etmemize yarıyor.



```python
df.info()
```

```python
df['Type 2'].value_counts(dropna=False)
```

value_count'la Type 2'nin içinde kaç kez ne oldugunu görüyorum, <br>
dropna=False ise NaN değer varsa onları da aynı şekilde göster demek. <br>
yani Type 2'nin içinde 386 tane NaN değer varmış.

```python
data1=df #aslında copy kullanmalıydık şuan ikisi de aynı yerde tutuldugu için yaptıgım degisiklik ikisini de etkileyecek.
data1.dropna(subset=["Type 2"],inplace=True) #nan value'ları drop ettik
```

inplace=True ise drop edilenler, veri setinden kalıcı olarak çıkartılır demektir.


```python
assert 1==1 #hiçbir şey döndürmedi çünkü doğru.
```

```python
assert 1==2 #yanlış olduğu için hata verdi.
```

```python
assert df['Type 2'].notnull().all() #error döndürmedi çünkü nan değer yok.
```

```python
df["Type 2"].fillna('empty',inplace=True) #nan olan yerleri empty ile doldurmak için kullanılır.
```

```python
assert df["Type 2"].notnull().all()
```
