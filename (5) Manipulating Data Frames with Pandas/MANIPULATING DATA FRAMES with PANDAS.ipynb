<h3>PIVOTING DATA FRAMES</h3> <br>
<li>pivoting: reshape tool</li>

```python
dic={"treatment":["A","A","B","B"],"gender":["F","M","F","M"],"response":[10,45,5,9],"age":[15,4,72,65]}
df=pd.DataFrame(dic)
df
```

bir data frame oluşturduk feature olarak tedavi, cinsiyet, tedaviye cevap ve yaş ekledik.

```python
#pivoting
df.pivot(index="treatment",columns="gender",values="response")
```

<strong>pivot</strong> ile index'i treatment, feature'ları gender verileri ise response yaptık.

<h3>STACKING and UNSTACKING DATAFRAME</h3> <br>
birden fazla indexe sahip data frame'leri unstack yapabiliyoruz.(indexlerinden bir tanesinde kurtulabiliyoruz.) <br>
level: unstack yapılacak index <br>
swaplevel: change inner and outer level index position

```python
df1=df.set_index(["treatment","gender"]) 
df1
```

veri setimdeki outer index:treatment, inner index: gender olsun dedik. 

```python
df1.unstack(level=0) #treatment artık index değil
```

```python
df1.unstack(level=1)  #gender index değil.
```

indexlerin yerlerini değiştirmek için yani inner ve outer'ın yerini:

```python
df2=df1.swaplevel(0,1)
df2
```

artık veri setimdeki outer index:gender, inner index: treatment oldu.

<h3>MELTING DATA FRAMES</h3> <br>

<li>Reverse of pivoting</li>

```python
pd.melt(df,id_vars="treatment",value_vars=["age","response"])
```

<strong>id_vars</strong> : treatment yani treatment sabit kalsın, <br>
<strong>value_vars</strong>:variable age ve response olsun value ise bunların tablodaki değerleri olsun.

```python
df.groupby("treatment").mean() #treatment'a göre grupla ve bunun ortalamasını al.
```
